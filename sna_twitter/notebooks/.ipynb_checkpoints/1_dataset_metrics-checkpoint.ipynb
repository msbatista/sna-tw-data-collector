```python
import urllib.parse
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from pymongo import MongoClient
from matplotlib.dates import DateFormatter

%matplotlib inline
sns.set(style="whitegrid")
```

```python
MONGO_PROCESSED_TWEETS = 'processed_tweets'
MONGO_BASE_STRING_CONN = 'mongodb://{}:{}@localhost:27017/'
MONGO_USERNAME = 'foo'
MONGO_PASSWORD = 'bar'

class MongoHandler:
    def __init__(self):
        self._client = MongoHandler._get_mongo_str_connection(user=MONGO_USERNAME, password=MONGO_PASSWORD)

    def get_collection(self, database: str, collection: str):
        try:
            return self._client[database][collection]
        except Exception as e:
            print(e)

    @staticmethod
    def _get_mongo_str_connection(user: str, password: str) -> MongoClient:
        parsed_username = urllib.parse.quote_plus(user)
        parsed_password = urllib.parse.quote_plus(password)

        return MongoClient(MONGO_BASE_STRING_CONN.format(parsed_username, parsed_password))
```

Setting connection with Mongo Database

```python
MONGO_DB = 'tweetsdb'
MONGO_COLLECTION = 'covid'

mongoHandler = MongoHandler()
tweets_collection = mongoHandler.get_collection(database=MONGO_DB, collection=MONGO_COLLECTION)
```

```python
df = pd.DataFrame(list(tweets_collection.find()))
df['created_at'] = pd.to_datetime(df['created_at'])
```

```python
df['date'] = pd.to_datetime(df['created_at'].dt.strftime('%Y-%m-%d %H:00'), format='%Y-%m-%d %H:00')
df = df.groupby('date')
tweets_count = df['id'].count()

plt.figure(figsize=(20,6))

ax = sns.lineplot(
    data=tweets_count,
    palette="tab10",
    linewidth=1.5,
    marker='o'
)

date_form = DateFormatter('%Y-%m-%d %H:00')

ax.xaxis.set_major_formatter(date_form)
ax.set(xlabel='Period', ylabel='NÂº of tweets')
ax.set_title('Covid dataset - Tweets by hour')
```
